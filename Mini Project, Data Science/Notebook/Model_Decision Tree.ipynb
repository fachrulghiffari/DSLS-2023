```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
```

```python
data_smotetomek = pd.read_csv(r'..\Hasil Proses Data\data_smotetomek.csv')
data_smoteenn = pd.read_csv(r'..\Hasil Proses Data\data_smoteenn.csv')
data_test = pd.read_csv(r'..\Hasil Proses Data\data_test.csv')
```

```python
data_test
```

```python
x_test = data_test.drop('level', axis=1)
y_test = data_test['level']
```

## Data dengan kombinasi SMOTE-TOMEK

```python
x_train_smotetomek = data_smotetomek.drop('level', axis=1)
x_train_smotetomek.sample(6)
```

```python
y_train_smotetomek = data_smotetomek['level']
```

```python
dt = DecisionTreeClassifier()

# Define the hyperparameter search space
param_grid = {
    'max_depth': [10, 15, 20, 25, 50, 75],
    'min_samples_split': [10, 15, 20, 25, 50, 75],
    'criterion': ['gini', 'entropy']
}

# Define the hyperparameter search method
grid_search = GridSearchCV(dt, param_grid, cv=5)

# Fit the model with different hyperparameters
grid_search.fit(x_train_smotetomek, y_train_smotetomek)

# Select the hyperparameters that result in the best performance on the validation set
best_params = grid_search.best_params_

# Evaluate the final performance of the model on the test set using the selected hyperparameters
dt = DecisionTreeClassifier(**best_params)
dt.fit(x_train_smotetomek, y_train_smotetomek)

train_score = dt.score(x_train_smotetomek, y_train_smotetomek)
print("Accuracy on training set: {:.2f}%".format(train_score*100))

# Evaluate the final performance of the model on the test set using the selected hyperparameters
test_score = dt.score(x_test, y_test)
print("Accuracy on test set: {:.2f}%".format(test_score*100))
```
