```python
data_smotetomek = pd.read_csv(r'..\Hasil Proses Data\data_smotetomek.csv')
data_smoteenn = pd.read_csv(r'..\Hasil Proses Data\data_smoteenn.csv')
data_smotetomek_norus = pd.read_csv(r'..\Hasil Proses Data\data_smotetomek_norus.csv')
data_smoteenn_norus = pd.read_csv(r'..\Hasil Proses Data\data_smoteenn_norus.csv')
data_test = pd.read_csv(r'..\Hasil Proses Data\data_test.csv')
```

```python
x_test = data_test.drop('level', axis=1)
y_test = data_test['level']
```

```python
x_test.sample(6)
```

```python
y_test
```

```python
def RandomForest(x_train, y_train, x_test, y_test):
    rf = RandomForestClassifier()

    # Define the hyperparameter search space
    n_estimators = (10, 100)
    max_features = ['auto', 'sqrt']
    max_depth = (10, 110)
    min_samples_split = (2, 10)
    min_samples_leaf = (1, 4)
    bootstrap = [True, False]

    search_space = {
        'n_estimators': n_estimators,
        'max_features': max_features,
        'max_depth': max_depth,
        'min_samples_split': min_samples_split,
        'min_samples_leaf': min_samples_leaf,
        'bootstrap': bootstrap
    }

    # Define the hyperparameter search method
    rf_search = BayesSearchCV(
        rf, search_space, n_iter=40, cv=5, n_jobs=-1, random_state=42, verbose=1
    )

    rf_search.fit(x_train, y_train)
    # Select the hyperparameters that result in the best performance
    best_params = rf_search.best_params_

    # Evaluate the final performance of the model using the selected hyperparameters
    rf = RandomForestClassifier(**best_params)

    rf.fit(x_train, y_train)

    train_score = rf.score(x_train, y_train)
    print("Accuracy on training set: {:.2f}%".format(train_score*100))

    # Evaluate the final performance of the model on the test set using the selected hyperparameters
    test_score = rf.score(x_test, y_test)
    print("Accuracy on test set: {:.2f}%".format(test_score*100))
```

## Data dengan kombinasi SMOTE-TOMEK

```python
x_train_smotetomek = data_smotetomek.drop('level', axis=1)
x_train_smotetomek.sample(6)
```

```python
y_train_smotetomek = data_smotetomek['level']
```

```python
RandomForest(x_train_smotetomek, y_train_smotetomek, x_test, y_test)
```

## Data dengan kombinasi SMOTE-ENN
